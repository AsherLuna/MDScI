```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate
import math
from decimal import *
```

```python
rango=1000
x=[]
i=0
while i<rango:
    i=i+0.01
    x.append(i)
```

```python
#O(1)
matriz=[]
for j in x:
    y=1
    matriz.append(y)
```

```python
#O(log(n))
lgr=[]
for k in x:
    result=math.log(k)
    lgr.append(result)
    base=x
```

```python
plt.figure(figsize=(15,15))
plt.title("O(1) vs O(log (n)")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,matriz)
plt.plot(base,lgr)
plt.grid()
plt.show()
```

```python
#O(n)
matrix=[]
for j in x:
    matrix.append(j)
```

```python
#O(nlog(n))
logar=[]
for m in x:
    res=math.log(m)
    logar.append(res*m)
    bass=x
```

```python
plt.figure(figsize=(15,15))
plt.title("O(n) vs O(nlog (n)")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,matrix)
plt.plot(bass,logar)
plt.grid()
plt.show()
```

```python
#O(n^2)
cuadrado=[]
for f in x:
    cuadrado.append(f*f)
```

```python
#O(n^3)
basis=x
cubo=[]
for m in x:
    cubo.append(m*m*m)
```

```python
plt.figure(figsize=(15,15))
plt.title("O(n^2) vs O(n^3)")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,cuadrado)
plt.plot(basis,cubo)
plt.grid()
plt.show()
```

```python
rango=100
w=[]
j=0
while j<rango:
    j=j+1
    w.append(j)
```

```python
#n^n
pot=[]
c=0
for f in w:
    c=(f**f)
    pot.append(c)
```

```python
#n!
#eje=x
factori=[]
for m in w:
    factori.append(math.factorial(m))
```

```python
plt.figure(figsize=(15,15))
plt.title("n! vs n^n")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(w,pot)
plt.plot(w,factori)
plt.grid()
plt.show()
```
