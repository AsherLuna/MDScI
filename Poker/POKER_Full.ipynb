```python
import random
import unicodedata
from collections import Counter
```

```python
def juego():
    mano=[]
    palo=[]
    indx=[]
    while len(mano)<5:
        carta=(random.randint(1,13),random.randint(1,4))
        if carta not in mano:
            mano.append(carta)
    mano.sort()
    #print("Mano: ",mano)
    for cartas in mano:
        indx.append(cartas[0])
        palo.append(cartas[1])
    
    result="Juego"
    valor=0
    count=Counter(palo)
    conteo=Counter(indx)
    
    #index=[1,1,7,11,12]
    #stick=[2,2,2,2,2]
    #count=Counter(stick)
    #conteo=Counter(index)
    #print("Palo",stick)
    #print("Valor",index)
    #print(conteo)
    #print("len: ",len(conteo))
    
    #print("Carta Alta")
    result="Carta Alta"
    
    if len(conteo)==5:
        if len(count)==1:
            result="Color"
            valor+=7
        elif indx[0]+4==indx[4]:
            #print("Escalera")
            result="Escalera"
            valor+=9
            if len(count)==1:
                #print("Escalera de Color")
                result="Escalera de Color"
                valor+=16
        else :
            if len(count)==1:
                if indx[0]==1 and indx[1]+3==indx[4] and indx[4]==13:
                    #print("Escalera Imperial")
                    result="Escalera Imperial"
                    valor+=17
                else:
                    #print("Color")
                    result="Color"
                    valor+=7
    
    elif len(conteo)==4:
        #print("Par")
        result="Par"
        valor+=2
        
    
    elif len(conteo)==3:
        for C in conteo:
            if conteo[C]==2:
                #print("Par")
                result="Dos Pares"
                valor+=2
            elif conteo[C]==3:
                #print("Tercia")
                result="Tercia"
                valor+=3
    
    elif len(conteo)==2:
        for C in conteo:
            if conteo[C]==4:
                #print("Poker")
                result="Poker"
                valor+=8
            elif conteo[C]==2:
                #print("Full House")
                result="Full House"
                valor+=5
    
    #print("el valor final es: ",valor)
    #print("Tu mano es: ",result)
    #print("")
    return result
```

```python
registro=[]
for i in range (0,2000):
    value=juego()
    registro.append(value)
conjunto=Counter(registro)
conjunto
```
